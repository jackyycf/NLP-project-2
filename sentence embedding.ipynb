```python
#loading data
import pandas as pd
import numpy as np
Data = pd.read_csv('data.tsv', sep='\t',on_bad_lines='skip',low_memory=False)
```

```python
Data
```

```python
Data=Data.dropna()
data = Data[Data['is_duplicate'] == 1]
data=data.reset_index()
data=data[data['index']<100000]
```

```python
data
```

```python
def preprocess(data):
    df=data.copy()
    df['question2'] = df.question2.str.lower()
    df['question2'] = df.question2.str.replace(r'[^\w\s]+', '',regex=True)
    res=[]
    for i in range(len(df)):
        res.append(df['question2'][i])
    return res
    
dataset=preprocess(data)
```

```python
dataset
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('roberta-large-nli-stsb-mean-tokens')
```

```python
sentence_embeddings = model.encode(dataset)
```

```python
np.save(('./sentence_embeddings.npy'), sentence_embeddings)
```

```python
sentence_embeddings=np.load(('./sentence_embeddings.npy'),allow_pickle=True)
```

```python
queries=['What are the best websites for entrepreneurs?']
```

```python
query_embeddings = model.encode(queries)
```

```python
import scipy
#test for sentence embedding algorithm
print("Search Results")
number_top_matches = 5
for query, query_embedding in zip(queries, query_embeddings):
    distances = scipy.spatial.distance.cdist([query_embedding], sentence_embeddings, "cosine")[0]
    results = zip(range(len(distances)), distances)
    results = sorted(results, key=lambda x: x[1])
    print("Query:", query)
    print("\nTop {} most similar sentences in corpus:".format(number_top_matches))

    for idx, distance in results[0:number_top_matches]:
        print(dataset[idx])
```
